Aggregate function

```python
import subprocess
import os

class yolo:
  def __init__(self):
    subprocess.Popen(["cd", "/content/darknet"] ,shell=True, 
                                  stdout=subprocess.PIPE, stderr=subprocess.PIPE)
    
    return 
  def train(self, data = "data/yolo.data",cfg = "cfg/yolov4-custom.cfg",weights = "yolov4.conv.137"):
    #python generate.py
    print("Training")

    #./darknet detector train data/yolo.data cfg/yolov4-custom.cfg yolov4.conv.137 -dont_s
    cmd_list = ['./darknet', 'detector', 'train', data, cfg, weights,'-dont_s']
    result = subprocess.run(cmd_list, stdout=subprocess.PIPE, stderr = subprocess.PIPE)
    print(result.stderr.decode())
    print(result.stdout.decode())

    #python tensorboard_generate.py
    cmd_list = ['python', 'tensorboard.py']
    result = subprocess.run(cmd_list, stdout=subprocess.PIPE, stderr = subprocess.PIPE)
    print(result.stderr.decode())
    print(result.stdout.decode())
    return 
  def validate(self,data = "data/yolo.data", cfg = "cfg/yolov4-custom.cfg", weights = "/content/darknet/yolov4-custom_3000.weights"):
    print("Testing")

    #./darknet detector valid data/yolo.data cfg/yolov4-custom.cfg '/content/darknet/yolov4-custom_3000.weights' 
    cmd_list = ['./darknet', 'detector', 'valid', data, cfg, weights]
    result = subprocess.run(cmd_list, stdout=subprocess.PIPE, stderr = subprocess.PIPE)
    print( result.stderr.decode())
    print (result.stdout.decode())
    #python coco_eval.py
    result = subprocess.run(['python','coco_eval.py'], stdout=subprocess.PIPE, stderr = subprocess.PIPE)
    print("COCO Evaluation")
    print( result.stderr.decode())
    print (result.stdout.decode())
    return
  def test_one(self, data = "data/yolo.data", cfg = "cfg/yolov4-custom.cfg", weights = "/content/darknet/yolov4-custom_3000.weights", image = "data/11003-Ravi-Singh-Baroda-Gujarat-1.jpg", thresh = "0.40"):
    
    #./darknet detector test data/yolo.data cfg/yolov4-custom.cfg '/content/darknet/yolov4-custom_3000.weights' 'data/11003-Ravi-Singh-Baroda-Gujarat-1.jpg' -thresh 0.40
    cmd_list = ['./darknet', 'detector', 'test', data, cfg, weights, image, '-thresh',thresh]
    result = subprocess.run(cmd_list, stdout=subprocess.PIPE, stderr = subprocess.PIPE)
    print("Testing sample image")
    print(result.stderr.decode())
    print (result.stdout.decode())
    return
```
